```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
usr = '1'
```

```python
def ReadData(DataName):
    return "/Users/huang/Desktop/IDVR_final/"+DataName

def ReMargin(x, z):
    xmin = x.min()
    zmin = z.min()
    
    x -= xmin
    z -= zmin
    
    return x, z
```

```python
# For U1 Fire
os.chdir("/Users/huang/Desktop/IDVR_final/Data/Fire/U" + usr)
excep_count = 0
excep_list = []

for idx, csv in enumerate(os.listdir()):
    try:
        df = pd.read_csv(csv)
        x, z = ReMargin(df['X'], df['Z'])
    #     title = 'Data #' + str(idx)
    #     plt.title(title)
        plt.plot(x, z, color='k')
        plt.axis('off')
        print(idx)
        imgName = 'U' + usr + '_' + str(idx-excep_count)
        imgPath = '/Users/huang/Desktop/IDVR_final/img/Fire/U' + usr + '/' + imgName + '.png'
        plt.savefig(imgPath)
        plt.show()
    except:
        print("EXCEPT")
        excep_count += 1
        excep_list.append(idx)
        
print("Fail :", excep_count)
print(excep_list)
```
